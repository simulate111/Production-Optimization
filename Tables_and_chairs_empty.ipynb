```python
import pyomo.environ as pyo #do not first import with star for this to be meaningful
m=pyo.ConcreteModel()

m.x1=pyo.Var(domain=pyo.NonNegativeReals)
m.x2=pyo.Var(domain=pyo.NonNegativeReals)

m.o=pyo.Objective(expr=5 * (100 * m.x1 + 60 * m.x2), sense=pyo.maximize)

m.c1=pyo.Constraint(expr=5 * m.x1 + 2 * m.x2 <= 27)
m.c2=pyo.Constraint(expr=4 * m.x1 + 3 * m.x2 <= 25)
m.c3=pyo.Constraint(expr=3 * m.x1 + 4 * m.x2 <= 20)

#Specify solver
solver = pyo.SolverFactory('cbc',executable='/bin/cbc')

#Verbous solving
results = solver.solve(m,tee=True)
print("Profit per week", m.o.expr())
print("Tables per day", m.x1.value,"chairs per day",m.x2.value)
```

Program in compact form

$\begin{align}
\max 5\sum_{p \in P} c_p x_p \end{align}$

s.t.


$\begin{align}
\sum_{p \in P} t_{ps}x_p \le a_s \;\; \forall s \in S \end{align}  $

$S=\{saw,sand,assemble\},P=\{tables,chairs\}$

$x_p \in \mathbb{Z}_{\ge 0} \;\; \forall p \in P$

```python
#GEmini
import pyomo.environ as pyo

m = pyo.ConcreteModel()

# Define Sets
m.P = pyo.Set(initialize=['tables', 'chairs'])
m.S = pyo.Set(initialize=['saw', 'sand', 'assemble'])

# Define Parameters
c = {'tables': 100, 'chairs': 60}
a = {'saw': 27, 'sand': 25, 'assemble': 20}
t = {
    ('tables', 'saw'): 5, ('chairs', 'saw'): 2,
    ('tables', 'sand'): 4, ('chairs', 'sand'): 3,
    ('tables', 'assemble'): 3, ('chairs', 'assemble'): 4
}

# Define Variables
m.x = pyo.Var(m.P, domain=pyo.NonNegativeIntegers)

# Define Objective Function
m.objective = pyo.Objective(
    expr=5 * sum(c[p] * m.x[p] for p in m.P),
    sense=pyo.maximize
)

# Define Constraints
def capacity_rule(model, s):
    return sum(t[(p, s)] * model.x[p] for p in model.P) <= a[s]
m.capacity = pyo.Constraint(m.S, rule=capacity_rule)

# Specify solver
solver = pyo.SolverFactory('cbc', executable='/bin/cbc')

# Solve the model
results = solver.solve(m, tee=True)

# Print results
print(f"\nProfit per week: {m.objective.expr()}")
for p in m.P:
    print(f"{p.capitalize()} per day: {m.x[p].value}")
```
