Most of the example code in this course is written using list comprehension: We can make lists and dictionaries using python.

````python
n=[2*n for n in range(1,6)] #fill a list with the expression 2*n where n i 1..5
print('The list:',n) #the result printed
print('The list in a one-line call:',[2*n for n in range(1,6)]) #sending the code for making the list to the print-function
print('The sum of the items in the list:',sum([2*n for n in range(1,6)])) #another function
print("The sum of the items in the list called 'n':",sum(n))
print('The sum of the items in the list generated using a list:',sum([2*n for n in [0,1,2,3,4,5]])) #here we do not use range(6) but define a list with he same items
print('The list from a list:',[n for n in n]) #first n is the result expression, second n is the iterator and third n is an iterable, i.e. a list called n we have created earlier
````

We will work with sets. We can use the keyword "in" to test if a item is in a set. Here the set is populated with pair of numbers. Using () gives a tuple. A tuple cannot be manipulated or updated. We will be able to generate Pyomo-sets form list, so we can alternatively create lists. Lists can contain duplicates of an item.

````python
n = {(i,j) for i in range(5) for j in range(5) if i < 3 and j >3}
print(n)
m = {(i,j) for i in range(5) for j in range(5) if (i,j) not in n} #set with tuples, the tuples must not be in n
print(m)
m = [[i,j] for i in range(5) for j in range(5) if (i,j) in n] #list with lists, the tuples needs to be in n
print(m)
m = [[i,j] for i in range(5) for j in range(5) if [i,j] in m] #we check if a list is in the list m
print(m)
print('A set',{1,2,3,3,3})
print('A list',[1,2,3,3,3])
````

In python everything is pointers do adresses so making copies are usually "shallow copies". This will usually not be a concern for us, since we are writing scripts most of the time. However, be careful when copying

````python
a = [1,2,3]
print('a',a)
b = a
print('b',b)
a.append(4) #we add a 4 to the list a
print('b',b) #but b is a!
````

````python
a = {'a':1,'b':2}
b = a
#This creates a new dictionary and copies from a
c = {}
for k in a.keys():
  c[k] = a[k]
#similarly using unpacking
d = {**a}
#or with dictionary compehension
e = {k:i for (k,i) in a.items()}
a['a'] = 3


print('a=',a,'b=',b,'c=',c,'d=',d,'e=',e)
````

Something on formating. Until now the output has been unformated. There are several way to format numbers and strings in the output in python. The most recently added formating is the f-format. Here you put the variable in a curly bracket and add formating information.

````python
pi = 3.14159654
print('The value of pi is', f"{pi:.2f}") #rounded two decimals
print('The value of pi is', f"{pi:7.4f}") #total seven positions, including '.' and two decimals
#also possible (using f-formating)
print(f"The value of pi is {pi:.2f}")
print(f"The value of pi is {pi:7.4f} or rounded to an integer {pi:.0f}")
````

It is a good practice to be restricitve on importing since you can mess things up. For example, generating random numers can be done with random using the module random or numpy. If you use "star import" you will not know which you are calling.

````python
#To be able to generate a random number (the module is called random and the function we wanted is called random we use an alias we like)
from random import random as rnd
print(rnd())
#Without an alias
from random import random
print(random())
#Or importing the whole packeged
import random
print(random.random()) #we now need to add the package name to to call
#Finally "star import", i.e. importing to the global name space. We can now call all functions in random with just there names
#(and hope there) are no other functions with the same name
from random import *
print(random())
print(randint(1,100)) #with starimport we can call randint that gives a random integer in the range we specify
print([randint(1,100) for i in range(randint(1,10))])
````

In scheduling we may want to deal with time and dates. For this the datetime module is very convenient.

````python
import datetime
#getting and printing the current time and date
now = datetime.datetime.now()
print("This is it!", now)
#adding 3 days, 2 hour and 37 minutes
time_to_add = datetime.timedelta(days=3, hours=2, minutes=37)
print("After 3 days, 2 hour and 37 minutes we have ",now+time_to_add)
#to get the year only
current_year = now.timetuple()[0] #this is a tuple with y,m,d,h,min,s
print("The year is", current_year)
#we can define a time point

christmas = datetime.datetime(current_year,12,24,18)
print("Only ",(christmas-now).total_seconds(), "seconds to Christmas dinner!")
````

Two examples of plotting. Takes a while to learn the syntax, but useful when illustrating results. See https://matplotlib.org/stable/plot_types/index.html for more info and examples.

````python
import math
import matplotlib.pyplot as plt #ploting similar to matlab
x = [0.01*i for i in range(615)]
y1 = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
plt.plot(x,y1,x,y2)
plt.show()
````

````python
#more advance plotting using a slighly different approach
#figure is the whole plot are whereas axes is waht you normally would call a plot
#This is ilke subplots in matlab, but here with only one
fig, ax = plt.subplots()
line1,= ax.plot(x, y1, 'g--') #ax.legend needs a tuple, this is a tuple with one element
line2,= ax.plot(x, y2,'c-.')
ax.set_xlabel('$x$')
ax.set_ylabel('$sin(\pi), cos(\pi)$')
ax.legend([line1,line2,],['$\dot{m}$','$m_{fix}$'])
ax.set_title('Not course related?')
plt.show()
````
